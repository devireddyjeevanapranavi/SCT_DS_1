```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r"C:\Users\devireddy ramreddy\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_130083\API_SP.POP.TOTL_DS2_en_csv_v2_130083.csv"

data = pd.read_csv(file_path, skiprows=4)
print(data.head())
```

```python
pop_2022 = data[['Country Name', '2022']].dropna()
print(pop_2022.head())
```

```python
top10 = pop_2022.sort_values('2022', ascending=False).head(10)
print(top10)
```

```python
plt.figure(figsize=(10,6))
plt.bar(top10['Country Name'], top10['2022'] / 1e9, 
        color=['#f7c948', '#58a4b0', '#d17b88', '#6b705c', '#81b29a', 
               '#e07a5f', '#3d405b', '#f2cc8f', '#118ab2', '#ef476f'])

plt.title("Top 10 Most Populous Countries (2022)", fontsize=14, fontweight='bold')
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population (in Billions)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(pop_2022['2022'] / 1e6, bins=20, color='#4e79a7', edgecolor='black')

plt.title("Distribution of Countries by Population (2022)", fontsize=14, fontweight='bold')
plt.xlabel("Population (in Millions)", fontsize=12)
plt.ylabel("Number of Countries", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
india_data = data[data['Country Name'] == 'India'].iloc[0, 4:-1]  # all year columns except metadata

years = india_data.index.astype(int)
population = india_data.values.astype(float) / 1e9  # Convert to billions

plt.figure(figsize=(10,6))
plt.plot(years, population, marker='o', color='#007acc')
plt.title("India's Population Growth (19602024)", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Population (in Billions)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
